```python
import scanpy as sc
adata_sc = sc.read_h5ad("../Step04_Clustering.h5ad")
```

```python
annotation_dict = {}

annotation_dict['0'] = "Neutrophils"
annotation_dict['1'] = "RBC"
annotation_dict['2'] = "Neutrophils"
annotation_dict['3'] = "PB"
annotation_dict['4'] = "CD16 Monocyte"
annotation_dict['5'] = "PB"
annotation_dict['6'] = "NK cells"
annotation_dict['7'] = "Platelet"
annotation_dict['8'] = "CD16 Monocyte"
annotation_dict['9'] = "T cells"
annotation_dict['10'] = "RBC"

annotation_dict
adata_sc.obs['cell_type'] = [annotation_dict[clust] for clust in adata_sc.obs['leiden_0.2']]
```

```python
import pandas as pd
adata = adata_sc
cell_proportion_df = pd.crosstab(adata.obs['Group'],adata.obs['cell_type'], normalize='columns').T.plot(kind='bar', stacked=True, legend=False).get_figure()
```
